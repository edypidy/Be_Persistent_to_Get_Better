# Flip img

```python
from PIL import Image

img_path = '/content/drive/MyDrive/images/doge.jpg'
img = Image.open(img_path)
flip_lr_img = img.transpose(Image.FLIP_LEFT_RIGHT)
flip_ud_img = img.transpose(Image.FLIP_TOP_BOTTOM)

figure, axes = plt.subplots(1, 3, figsize=(18, 4))
axes[0].imshow(img)
axes[1].imshow(flip_lr_img)
axes[2].imshow(flip_ud_img)
```

```python
from PIL import Image
from PIL import ImageFilter

img_path = '/content/drive/MyDrive/images/doge.jpg'
img = Image.open(img_path)

blur_img = img.filter(ImageFilter.BLUR) # fixed value
blur_img1 = img.filter(ImageFilter.BoxBlur(10))
blur_img2 = img.filter(ImageFilter.GaussianBlur(10))

figure, axes = plt.subplots(1, 4, figsize=(18, 4))
axes[0].imshow(img)
axes[1].imshow(blur_img)
axes[2].imshow(blur_img1)
axes[3].imshow(blur_img2)
```

# Combine img

```python
from PIL import Image

img_path = '/content/drive/MyDrive/images/doge.jpg'
img = Image.open(img_path)

new_img = Image.new('RGB', (1000, 750), color=123456)
new_img.paste(img.resize((100, 75)), (10,10))
new_img.paste(img.resize((200, 150)), (100 + 10, 10))
new_img.paste(img.resize((300, 225)), (10, 150 + 10))
new_img.paste(img.resize((400, 300)), (300 + 10, 150 + 10))
new_img.paste(img.resize((500, 375)), (10, 300 + 150 + 10))
new_img.paste(img.resize((600, 450)), (500 + 10, 300 + 150 + 10))

plt.imshow(new_img)
```

# Cf) img to pixel values & Get RGB color of pixel we want

```python
from PIL import Image
import numpy as np

img_path = '/content/drive/MyDrive/images/doge.jpg'
img = Image.open(img_path)

pixel_list = list(img.getdata())
print(pixel_list)
np.savetxt('pixel_type_data.txt', pixel_list, fmt='%d', delimiter='')
```

```python
from PIL import Image

img_path = '/content/drive/MyDrive/images/doge.jpg'
img = Image.open(img_path)

rgb_img = img.convert('RGB')
pixel_point = (500, 500)

tuple_item = rgb_img.getpixel(pixel_point)
print('tuple item : ', tuple_item)
```

```python
r, g, b = tuple_item
print(f'R : {r}, G : {g}, B : {b}')
```
